````python
# Define os valores e os pesos
valores = [5, 12, 20, 15]
pesos = [1, 2, 3, 4]

# Calcula a média ponderada
media_ponderada = sum(valor * peso for valor, peso in zip(valores, pesos)) / sum(pesos)

# Imprime a média ponderada
print(f"A média ponderada é {media_ponderada}.")
````

#### Questão 10
Um programa deve ser escrito para ler dois números e, em seguida, perguntar à pessoa usuária qual operação ele deseja realizar. O resultado da operação deve incluir informações sobre o número - se é par ou ímpar, positivo ou negativo e inteiro ou decimal.

````python
# Solicita dois números
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

# Solicita a operação a ser realizada
operacao = input("Digite a operação que deseja realizar (+, -, *, /): ")

# Realiza a operação e armazena o resultado
if operacao == "+":
    resultado = numero1 + numero2
elif operacao == "-":
    resultado = numero1 - numero2
elif operacao == "*":
    resultado = numero1 * numero2
elif operacao == "/":
    if numero2 != 0:
        resultado = numero1 / numero2
    else:
        print("Erro: Divisão por zero não é permitida.")
        resultado = None
else:
    print("Erro: Operação inválida.")
    resultado = None

# Exibe o resultado da operação
if resultado is not None:
    if resultado % 2 == 0:
        par_ou_impar = "par"
    else:
        par_ou_impar = "ímpar"
    if resultado > 0:
        positivo_ou_negativo = "positivo"
    elif resultado < 0:
        positivo_ou_negativo = "negativo"
    else:
        positivo_ou_negativo = "zero"
    if resultado.is_integer():
        inteiro_ou_decimal = "inteiro"
    else:
        inteiro_ou_decimal = "decimal"
    print(f"O resultado da operação é {resultado}, que é {par_ou_impar}, {positivo_ou_negativo} e {inteiro_ou_decimal}.")
````

#### Questão 11
Escreva um programa que peça à pessoa usuária três números que representam os lados de um triângulo. O programa deve informar se os valores podem ser utilizados para formar um triângulo e, caso afirmativo, se ele é equilátero, isósceles ou escaleno. Tenha em mente algumas dicas:
- Três lados formam um triângulo quando a soma de quaisquer dois lados for maior que o terceiro;
- Triângulo Equilátero: três lados iguais;
- Triângulo Isósceles: quaisquer dois lados iguais;
- Triângulo Escaleno: três lados diferentes;


````python
# Solicita os três lados do triângulo
lado1 = float(input("Digite o primeiro lado do triângulo: "))
lado2 = float(input("Digite o segundo lado do triângulo: "))
lado3 = float(input("Digite o terceiro lado do triângulo: "))

# Verifica se os lados formam um triângulo
if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    # Verifica se o triângulo é equilátero
    if lado1 == lado2 == lado3:
        tipo_triangulo = "equilátero"
    # Verifica se o triângulo é isósceles
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        tipo_triangulo = "isósceles"
    # Verifica se o triângulo é escaleno
    else:
        tipo_triangulo = "escaleno"
    print(f"Os lados formam um triângulo {tipo_triangulo}.")
else:
    print("Os lados não formam um triângulo.")
````

Para o diesel, se a quantidade comprada for até 15 litros, o desconto será de 3% por litro. Caso contrário, será de 5% por litro. O preço do litro de diesel é R\$ 2,00 e o preço do litro de etanol é R\$ 1,70. Escreva um programa que leia a quantidade de litros vendidos e o tipo de combustível (E para etanol e D para diesel) e calcule o valor a ser pago pelo cliente. Tenha em mente algumas dicas:

- O do valor do desconto será a multiplicação entre preço do litro, quantidade de litros e o valor do desconto.
- O valor a ser pago por um cliente será o resultado da multiplicação do preço do litro pela quantidade de litros menos o valor de desconto resultante do cálculo.


````python
# Solicita a quantidade de litros e o tipo de combustível
litros = float(input("Digite a quantidade de litros comprados: "))
combustivel = input("Digite o tipo de combustível (E para etanol ou D para diesel): ").upper()
preco_litro = 1.7 if combustivel == "E" else 2.0

# Calcula o desconto
if litros <= 15:
    desconto = 0.02 * preco_litro if combustivel == "E" else 0.03 * preco_litro
else:
    desconto = 0.04 * preco_litro if combustivel == "E" else 0.05 * preco_litro

# Calcula o valor a ser pago
valor_a_pagar = litros * preco_litro - litros * preco_litro * desconto

# Exibe o resultado
print(f"O valor a ser pago pelo cliente é R${valor_a_pagar:.2f}.")
````

Em uma empresa de venda de imóveis você precisa criar um código que analise os dados de vendas anuais para ajudar a diretoria na tomada de decisão. O código precisa coletar os dados de quantidade de venda durante os anos de 2022 e 2023 e fazer um cálculo de variação percentual. A partir do valor da variação, deve ser enviada às seguintes sugestões:
- Para variação acima de 20%: bonificação para o time de vendas.
- Para variação entre 2% e 20%: pequena bonificação para time de vendas.
- Para variação entre 2% e -10%: planejamento de políticas de incentivo às vendas.
- Para bonificações abaixo de -10%: corte de gastos.


````python
# Solicita os dados de vendas anuais
vendas_2022 = float(input("Digite a quantidade de vendas do ano de 2022: "))
vendas_2023 = float(input("Digite a quantidade de vendas do ano de 2023: "))

# Calcula a variação percentual
variacao = ((vendas_2023 - vendas_2022) / vendas_2022) * 100

# Faz a análise da variação percentual
if variacao > 20:
    print("Bonificação para o time de vendas.")
elif variacao > 2:
    print("Pequena bonificação para o time de vendas.")
elif variacao > -10:
    print("Planejamento de políticas de incentivo às vendas.")
else:
    print("Corte de gastos.")

# Exibe o valor da variação percentual
print(f"A variação percentual entre os anos de 2022 e 2023 foi de {variacao:.2f}%.")
````

Escreva um programa que peça dois números inteiros e imprima todos os números inteiros entre eles.

````python
# Solicita dois números inteiros
numero_1 = int(input("Digite o primeiro número inteiro: "))
numero_2 = int(input("Digite o segundo número inteiro: "))

# Ordena os números em ordem crescente
if numero_1 > numero_2:
    numero_1, numero_2 = numero_2, numero_1

# Imprime todos os números inteiros entre os dois números
for numero in range(numero_1 + 1, numero_2):
    print(numero)
    
````

````python
# Define a lista de salarios
salarios = [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]

# Calcula o abono para cada salário
abonos = {salario: max(salario * 0.1, 200) for salario in salarios}

# Calcula o total de gastos com o abono
total_abono = sum(abonos.values())

# Conta quantos funcionários receberam o abono mínimo
funcionarios_minimo = sum(1 for salario in abonos.keys() if abonos[salario] == 200)

# Encontra o maior valor de abono fornecido
maior_abono = max(abonos.values())

# Exibe o total de gastos com o abono, a quantidade de funcionários que receberam o abono mínimo, e o maior valor de abono fornecido
print(f"Total de gastos com o abono: R$ {total_abono:.2f}")
print(f"{funcionarios_minimo} funcionários receberam o abono mínimo de R$ 200")
print(f"O maior valor de abono fornecido foi de R$ {maior_abono:.2f}")
````

#### Questão 14
Uma equipe de cientistas de dados está estudando a diversidade biológica em uma floresta. A equipe fez a coleta de informações sobre o número de espécies de plantas e animais em cada área da floresta e armazenaram essas informações em um dicionário. Nele, a chave descreve a área dos dados e os valores nas listas correspondem às espécies de plantas e animais nas áreas, respectivamente.
```
{'Área Norte': [2819, 7236],
 'Área Leste': [1440, 9492],
 'Área Sul': [5969, 7496],
 'Área Oeste': [14446, 49688],
 'Área Centro': [22558, 45148]}
```
Escreva um código para calcular a média de espécies por área e identificar a área com a maior diversidade biológica. Dica: use as funções built-in [`sum()`](https://docs.python.org/3/library/functions.html#sum) e [`len()`](https://docs.python.org/3/library/functions.html#len).


````python
# Define o dicionario das especies de dados
dados = {
    'Área Norte': [2819, 7236],
    'Área Leste': [1440, 9492],
    'Área Sul': [5969, 7496],
    'Área Oeste': [14446, 49688],
    'Área Centro': [22558, 45148]
}

# Calcula o total de espécies de plantas e animais
soma_plantas = sum(area[0] for area in dados.values())
soma_animais = sum(area[1] for area in dados.values())

# Calcula a média de espécies por área
media_especies = (soma_plantas + soma_animais) / 2

# Encontra a área com a maior diversidade biológica
area_maior_biodiversidade = max(dados, key=lambda area: (dados[area][0] + dados[area][1]))

# Exibe a média de espécies por área e a área com a maior diversidade biológica
print(f"A média de espécies por área é: {media_especies:.2f}")
print(f"A área com a maior diversidade biológica é: {area_maior_biodiversidade}")
````

#### Questão 15
O setor de RH da sua empresa te pediu uma ajuda para analisar as idades dos funcionários de 4 setores da empresa. Para isso, ele te forneceu os seguintes dados:
```
{'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}
``` 
Sabendo que cada setor tem 10 funcionários, construa um código que calcule a média de idade de cada setor, a idade média geral entre todos os setores e quantas pessoas estão acima da idade média geral.


````python
# Define o dicionario de idade por departamento
dados = {
    'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
    'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
    'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
    'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
}

# Calcula o total de funcionários
total_funcionarios = sum(len(setor) for setor in dados.values())

# Calcula a soma das idades de todos os funcionários
soma_idades = sum(sum(setor) for setor in dados.values())

# Calcula a idade média de cada setor
idade_media_setores = [sum(setor) / len(setor) for setor in dados.values()]

# Calcula a idade média geral entre todos os setores
idade_media_geral = soma_idades / total_funcionarios

# Conta quantas pessoas estão acima da idade média geral
acima_idade_media = sum(1 for setor in dados.values() if sum(setor) / len(setor) > idade_media_geral)

# Exibe a idade média de cada setor, a idade média geral entre todos os setores, e a quantidade de pessoas acima da idade média geral
print(f"A idade média de cada setor é: {idade_media_setores}")
print(f"A idade média geral entre todos os setores é: {idade_media_geral:.2f}")
print(f"{acima_idade_media} pessoas estão acima da idade média geral.")
````
