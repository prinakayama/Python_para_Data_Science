## a = Coeficiente angular
## n -> numero de elementos
## Y = Moscow
## X = datas


```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
(X ** 2).shape
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**Y)-np.sum(X)**2)
```

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X*b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x,y)
plt.plot(41.5, 41.5*a+b, '*r')
plt.plot(100, 100*a+b, '*r')
```
