```python
import pandas as pd
import requests

url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_hospedagem.json'
response = requests.get(url)

with open('dados_hospedagem.json', 'wb') as f:
    f.write(response.content)

df = pd.read_json('dados_hospedagem.json')
```

```python
df.head()
```

```python
#Transforma os dados em um formato de tbl com
df = pd.json_normalize(df['info_moveis'])
df
```

```python
#retirando listas
colunas = list(df.columns)
colunas
```

```python
#Desagrupar os dados
df = df.explode(colunas[3:])
df
```

```python
#Resetando os indices para ordenar os dados
df.reset_index(inplace = True, drop=True)
df.head()
```

```python
#tipos de dados
df.info()
```

```python
df['preco']
```

```python
# aplly lambda -> aplica uma funcao especifica a cada elemento (strip, apaga espacos em branco)
df['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
df['preco'] = df['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
import numpy as np
```

```python
# metodo astype converte o tipo
df['preco'] = df['preco'].astype(np.float64)
```

```python
df.info()
```

```python
#conversao de multiplas colunas
df[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
df[['taxa_deposito','taxa_limpeza']] = df[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
df[['taxa_deposito','taxa_limpeza']] = df[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
df.info()
```

```python
#analise textual
#Tokenizacao simples
df['descricao_local']
```

```python
df['descricao_local'] = df['descricao_local'].str.lower()
```

```python
df.head()
```

```python
df['descricao_local'][3169]
```

```python
df['descricao_local'] = df['descricao_local'].str.replace('[^a-zA-Z0-9\-]', '', regex=True)
```

```python
df['descricao_local'] = df['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
#metodo split separa cada elemento da lista

df['descricao_local'] = df['descricao_local'].str.split()
df.head()
```

```python
df['comodidades'] = df['comodidades'].str.replace('\{|}|\"','',regex = True)
```

```python
df['comodidades'] = df['comodidades'].str.split(',')
df.head()
```
